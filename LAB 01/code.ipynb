```python
import time
import random
import itertools
import numpy as np
```

```python
# function for the initial matrix, target matrix and matrix- characters
def parseInputs():

    # initial matrix
    gridCharacters = "12345678B"
    gridCharacters = ''.join(random.sample(gridCharacters,len(gridCharacters)))
    temp=gridCharacters[:]
    print("Initial Grid: ")
    print(np.reshape([*temp],(3,3)))

    # target matrix
    targetCharacters = "12345678B"
    temp=targetCharacters[:]
    print("Target Grid: ")
    print(np.reshape([*temp],(3,3)))

    # solvable characters
    solvableCharacters = "123B46758"

    return gridCharacters, targetCharacters, solvableCharacters
```

```python
# function to find out the initial x and y coordinates of the blank element
def initialBlankCoordinates(gridCharacters):

    for i in range(len(gridCharacters)):
        if gridCharacters[i] == 'B':
            return i

    return -1
```

```python
# accessing elements at (x-1,y), (x+1,y), (x,y-1) and (x,y+1) unless invalid
def goUp(idx):
    if idx > 2:
        return idx - 3
    return -1

def goDown(idx):
    if idx < 6:
        return idx + 3
    return -1

def goLeft(idx):
    if idx != 0 and idx != 3 and idx != 6:
        return idx -1
    return -1

def goRight(idx):
    if idx !=2 and idx != 5 and idx != 8:
        return idx + 1
    return -1
```

```python
# BFS function
def performBFS(gridCharacters, targetCharacters, idx):

    visitedDict = {}
    queue = []
    
    # to keep track of visited configurations
    visitedDict[gridCharacters] = 1
    # Storing in the form of a tuple
    queue.append((gridCharacters,idx))
  
    # counter variable 
    cld = 0

    while queue:
      # first element in queue-> m[0]: string for matrix configuration, m[1]: index
      m = queue.pop(0)
      
      # target configuration reached  
      if m[0] == targetCharacters:
        print('Finished in ' + str(cld))
        return True
        
      currentBlank = m[1]
    
      # if (x-1,y) is a valid move
      if goUp(currentBlank) != -1:
        cpy = m[0][:]
        newBlank = goUp(currentBlank)
        newBlankCharacter = cpy[newBlank]
        cpy = cpy[:currentBlank] + cpy[newBlank] + cpy[currentBlank+1:]
        cpy = cpy[:newBlank] + "B" + cpy[newBlank+1:]

        # checking if configuration is already present in visited   
        if visitedDict.get(cpy) == None:
          visitedDict[cpy] = 1
          cld = cld + 1
          queue.append((cpy,newBlank))
            
      # if (x+1,y) is a valid move
      if goDown(currentBlank) != -1:
        cpy = m[0][:]
        newBlank = goDown(currentBlank)
        newBlankCharacter = cpy[newBlank]
        cpy = cpy[:currentBlank] + cpy[newBlank] + cpy[currentBlank+1:]
        cpy = cpy[:newBlank] + "B" + cpy[newBlank+1:]

        # checking if configuration is already present in visited     
        if visitedDict.get(cpy) == None:
          visitedDict[cpy] = 1
          cld = cld + 1
          queue.append((cpy,newBlank))
            
      # if (x,y+1) is a valid move
      if goRight(currentBlank) != -1:
        cpy = m[0][:]
        newBlank = goRight(currentBlank)
        newBlankCharacter = cpy[newBlank]
        cpy = cpy[:currentBlank] + cpy[newBlank] + cpy[currentBlank+1:]
        cpy = cpy[:newBlank] + "B" + cpy[newBlank+1:]

        # checking if configuration is already present in visited     
        if visitedDict.get(cpy) == None:
          visitedDict[cpy] = 1
          cld = cld + 1
          queue.append((cpy,newBlank))

      # if (x,y-1) is a valid move
      if goLeft(currentBlank) != -1:
        cpy = m[0][:]
        newBlank = goLeft(currentBlank)
        newBlankCharacter = cpy[newBlank]
        cpy = cpy[:currentBlank] + cpy[newBlank] + cpy[currentBlank+1:]
        cpy = cpy[:newBlank] + "B" + cpy[newBlank+1:]
            
        # checking if configuration is already present in visited     
        if visitedDict.get(cpy) == None:
          visitedDict[cpy] = 1
          cld = cld + 1
          queue.append((cpy,newBlank))
             
    print('Finished in ' + str(cld))
    return False
```

```python
# DFS function
def performDFS(gridCharacters, targetCharacters, idx):

    visitedDict = {}
    stack = []
    
    # to keep track of visited configurations
    visitedDict[gridCharacters] = 1
    # Storing in the form of a tuple
    stack.append((gridCharacters,idx))
  
    # counter variable 
    cld = 0

    while stack:
      # first element in queue-> m[0]: string for matrix configuration, m[1]: index
      m = stack.pop()
      
      # target configuration reached  
      if m[0] == targetCharacters:
        print('Finished in ' + str(cld))
        return True
        
      currentBlank = m[1]
    
      # if (x-1,y) is a valid move
      if goUp(currentBlank) != -1:
        cpy = m[0][:]
        newBlank = goUp(currentBlank)
        newBlankCharacter = cpy[newBlank]
        cpy = cpy[:currentBlank] + cpy[newBlank] + cpy[currentBlank+1:]
        cpy = cpy[:newBlank] + "B" + cpy[newBlank+1:]

        # checking if configuration is already present in visited   
        if visitedDict.get(cpy) == None:
          visitedDict[cpy] = 1
          cld = cld + 1
          stack.append((cpy,newBlank))
            
      # if (x+1,y) is a valid move
      if goDown(currentBlank) != -1:
        cpy = m[0][:]
        newBlank = goDown(currentBlank)
        newBlankCharacter = cpy[newBlank]
        cpy = cpy[:currentBlank] + cpy[newBlank] + cpy[currentBlank+1:]
        cpy = cpy[:newBlank] + "B" + cpy[newBlank+1:]

        # checking if configuration is already present in visited     
        if visitedDict.get(cpy) == None:
          visitedDict[cpy] = 1
          cld = cld + 1
          stack.append((cpy,newBlank))
            
      # if (x,y+1) is a valid move
      if goRight(currentBlank) != -1:
        cpy = m[0][:]
        newBlank = goRight(currentBlank)
        newBlankCharacter = cpy[newBlank]
        cpy = cpy[:currentBlank] + cpy[newBlank] + cpy[currentBlank+1:]
        cpy = cpy[:newBlank] + "B" + cpy[newBlank+1:]

        # checking if configuration is already present in visited     
        if visitedDict.get(cpy) == None:
          visitedDict[cpy] = 1
          cld = cld + 1
          stack.append((cpy,newBlank))

      # if (x,y-1) is a valid move
      if goLeft(currentBlank) != -1:
        cpy = m[0][:]
        newBlank = goLeft(currentBlank)
        newBlankCharacter = cpy[newBlank]
        cpy = cpy[:currentBlank] + cpy[newBlank] + cpy[currentBlank+1:]
        cpy = cpy[:newBlank] + "B" + cpy[newBlank+1:]
            
        # checking if configuration is already present in visited     
        if visitedDict.get(cpy) == None:
          visitedDict[cpy] = 1
          cld = cld + 1
          stack.append((cpy,newBlank))
             
    print('Finished in ' + str(cld))
    return False
```

```python
# calling function for the initial matrix, target matrix and matrix- characters
gridCharacters, targetCharacters, solvableCharacters= parseInputs()

# index of current blank element
idx  = initialBlankCoordinates(gridCharacters)

print("BFS")
print(performBFS(gridCharacters, targetCharacters, idx))
print("DFS")
print(performDFS(gridCharacters, targetCharacters, idx))
```
